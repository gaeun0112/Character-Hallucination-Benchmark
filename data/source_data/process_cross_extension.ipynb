```python
import json

with open('./cise_cross_choices_version3_check_fin.extended_only.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

print(data)
```

```python
for country in data:
    for character in data[country]:
        for idx, row in enumerate(data[country][character]):
            real_answer = data[country][character][idx]['Answer']
            cannot_answer = data[country][character][idx]['Incorrect Answer 9']

            data[country][character][idx]['Answer'] = cannot_answer
            data[country][character][idx]['Incorrect Answer 9'] = real_answer
```
