```python
import json

def read_json(file_path):

    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        return data
```

```python
def save_json(data, file_path):
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
cross_data = read_json('./cise_cross_choices_version3_check_fin.json')
cross_factual_data = read_json('./cross_factual_choices_check_fin.json')
```

```python
cross_data
```

```python
cross_factual_data
```

```python
for country in cross_data:
    for character in cross_data[country]:
        cross_data[country][character] += cross_factual_data[country][character]
```

```python
save_json(data=cross_data, file_path="./cross_choices_check_fin.json")
```
