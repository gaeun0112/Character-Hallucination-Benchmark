```python
import json 

with open('./cultural_choices_descriptive.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
for country in data:
    for obj in data[country]:
        print(country)
        print("===")
        print(obj['country_choices'])
        for ctr in obj['country_choices']:
            if ctr in ['US', 'UK']:
                print(ctr)
        break
```
